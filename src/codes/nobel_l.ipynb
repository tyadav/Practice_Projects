```python
from datetime import date 
  
def calculateAge(birthDate): 
    today = date.today() 
    age = today.year - birthDate.year - ((today.month, today.day) < 
         (birthDate.month, birthDate.day)) 
  
    return age 
```

```python
print(calculateAge(date(1971, 7, 23)), "years") 
```
